# Reflection Oporator

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------------------
# Reflection axis |v>
# -------------------------------------------------

theta = np.pi / 4  # 45 degrees
v = np.array([np.cos(theta), np.sin(theta)])
v = v / np.linalg.norm(v)

# Reflection operator
R = 2 * np.outer(v, v) - np.eye(2)

# Original state
psi = np.array([0.9, 0.25])
psi = psi / np.linalg.norm(psi)

# Reflected state
psi_reflected = R @ psi

plt.figure(figsize=(7,7))

# Reflection axis
t = np.linspace(-1.3, 1.3, 100)
plt.plot(t*v[0], t*v[1],
         'k--',
         linewidth=2,
         label=r'Reflection Axis $|v\rangle$')

# Original state
plt.quiver(0,0,
           psi[0],psi[1],
           angles='xy',
           scale_units='xy',
           scale=1,
           color='royalblue',
           label=r'Original State $|\psi\rangle$')

# Reflected state
plt.quiver(0,0,
           psi_reflected[0],
           psi_reflected[1],
           angles='xy',
           scale_units='xy',
           scale=1,
           color='crimson',
           label=r'Reflected State $R|\psi\rangle$')

# Dotted line connecting them
plt.plot([psi[0], psi_reflected[0]],
         [psi[1], psi_reflected[1]],
         linestyle=':',
         color='gray')

# Labels
plt.text(v[0]*1.08, v[1]*1.08,
         r'$|v\rangle$',
         fontsize=14)

plt.text(psi[0]*1.08, psi[1]*1.08,
         r'$|\psi\rangle$',
         color='royalblue',
         fontsize=14)

plt.text(psi_reflected[0]*1.08,
         psi_reflected[1]*1.08,
         r'$R|\psi\rangle$',
         color='crimson',
         fontsize=14)

plt.axhline(0,color='black')
plt.axvline(0,color='black')

plt.xlim(-1.2,1.2)
plt.ylim(-1.2,1.2)
plt.gca().set_aspect('equal')

plt.xlabel("x")
plt.ylabel("y")
 
plt.title("Geometric Interpretation of the Reflection Operator |v⟩ = (1/√2)[1, 1]")
plt.grid(True)
plt.legend(loc='upper left')

plt.show()
```
