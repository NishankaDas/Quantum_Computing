```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_walk(initial=0, steps=100, p=0.5):
    """
    Simulate one classical random walk.

    Parameters
    ----------
    initial : int
        Initial position.
    steps : int
        Number of steps.
    p : float
        Probability of moving right.

    Returns
    -------
    int
        Final position.
    """

    position = initial

    for _ in range(steps):
        if np.random.rand() < p:
            position += 1
        else:
            position -= 1

    return position
```

```python
def run(trials=10000, steps=100, initial=0, p=0.5):

    final_positions = [random_walk(initial, steps, p)
                       for _ in range(trials)]

    positions, counts = np.unique(final_positions,
                                  return_counts=True)

    probability = counts / trials

    plt.figure(figsize=(10,5))
    plt.bar(positions,
            probability,
            width=1.8,
            color="steelblue",
            edgecolor="black")

    plt.xlabel("Position")
    plt.ylabel("Probability")
    plt.title(f"Classical Random Walk ({steps} Steps)")
    plt.grid(alpha=0.3)
    plt.show()
```

```python
run(trials=50000, steps=100)
```
